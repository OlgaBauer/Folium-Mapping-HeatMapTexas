```python
from folium.plugins import HeatMap
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
drought_heatmap_df = pd.read_csv(r"C:\Python\ob\Folium Mapping Drought\ob_drought_heatmap.csv")
drought_heatmap_df.dropna(inplace=True)
drought_heatmap_df.head()
```

```python
import branca.colormap
from collections import defaultdict
```

```python
from branca.element import Element
from branca.colormap import LinearColormap
```

```python
linear = branca.colormap.LinearColormap(['green', 'yellow', 'red'],
                           vmin=1, vmax=15)
linear
```

```python
hm = folium.Map(location=[28, -90], width=1900, height=940,
                 zoom_start=6.3, 
                 control_scale=True)


colormap = branca.colormap.LinearColormap(['green', 'yellow', 'red'], index=[5,10,14], 
                           vmin=2, vmax=14)
colormap.caption = ('Consecutive weeks of drought')
colormap.add_to(hm) #add color bar at the top of the map

#popup
for i,row in drought_heatmap_df.iterrows():
    #Setup the content of the popup
    iframe = folium.IFrame(str(row["Weeks"]) + ' weeks of drought')
    
    #Initialise the popup using the iframe
    popup = folium.Popup(iframe, min_width=300, max_width=300)
#marker
    #Add each row to the map
    folium.Marker(location=[row['Latitude'],row['Longitude']], 
                  icon=folium.Icon(color='red', icon='info', prefix='fa'),
                  popup = popup)

HeatMap(drought_heatmap_df,
        min_opacity=0.4,
        blur = 18
               ).add_to(folium.FeatureGroup(name='Heat Map').add_to(hm))
folium.LayerControl().add_to(hm)

hm.save("index.html")

hm
```
